# To create a vision program to find histogram value and display histograph of a grayscale and color image.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2gray
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image = io.imread("/content/apple2.jpg")
gray = rgb2gray(image)
gray = (gray*255).astype(np.uint8)
```

```python
plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title("Original Color Image")
plt.axis("off")
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.title("Color Image Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Count")

plt.plot(cv2.calcHist([image], [0], None, [256], [0,256]), color='red', label='Red')
plt.plot(cv2.calcHist([image], [1], None, [256], [0,256]), color='green', label='Green')
plt.plot(cv2.calcHist([image], [2], None, [256], [0,256]), color='blue', label='Blue')

plt.legend()

```

```python
plt.subplot(2, 2, 2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
plt.subplot(1,2,2)
plt.title("Gray Image Histogram")
plt.xlabel("pixel value")
plt.ylabel("count")
plt.hist(gray.ravel(), bins= 256, range= [0,256],color = 'gray',label ='gray')
plt.legend()
plt.tight_layout()
```
